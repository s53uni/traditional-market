## 부산 전통시장 활성화 방안(설문조사) 시각화

### 라이브러리 및 한글 처리

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rc("font", family = "Malgun Gothic")
```

### 데이터 불러오기

```python
df_2017 = pd.read_csv("./data/재래시장_활성화_방안_2017.csv", encoding="euc-kr")
df_2019 = pd.read_csv("./data/재래시장_활성화_방안_2019.csv", encoding="euc-kr")
df_2021 = pd.read_csv("./data/재래시장_활성화_방안_2021.csv", encoding="euc-kr")
```

```python
df_2021
```

### 2021년 연령별 활성화 방안 : df_2021_age

```python
# 2021년 데이터에서 연령별 데이터만 추출
df_2021_age = df_2021[(df_2021["구분별(1)"] == "연령별")].drop("구분별(1)", axis=1)
df_2021_age.reset_index(drop=True, inplace=True)

# 컬럼명 변경하기
df_2021_age.rename(columns={"구분별(2)":"연령대"}, inplace=True)
df_2021_age
```

```python
df_2021_age.to_csv("./data_new/부산_2021년_연령대별_전통시장_활성화_방안.csv", encoding="euc-kr", index=False)
```

### 2019년 연령별 활성화 방안 : df_2019_age

```python
df_2019_age = df_2019[(df_2019["특성별(1)"] == "연령별")].drop("특성별(1)", axis=1).reset_index(drop=True)
df_2019_age.rename(columns={"특성별(2)":"연령대"}, inplace=True)
df_2019_age
```

```python
df_2019_age.to_csv("./data_new/부산_2019년_연령대별_전통시장_활성화_방안.csv", encoding="euc-kr", index=False)
```

### 2017년 연령별 활성화 방안 : df_2017_age

```python
df_2017_age = df_2017[(df_2017["특성별(1)"] == "연령별")].drop("특성별(1)", axis=1).reset_index(drop=True)
df_2017_age.rename(columns={"특성별(2)":"연령대"}, inplace=True)
df_2017_age
```

```python
df_2017_age.to_csv("./data_new/부산_2017년_연령대별_전통시장_활성화_방안.csv", encoding="euc-kr", index=False)
```

### 2017,2019,2021년의 전체 전통시장 활성화 방안 비교

```python
df_2021_all = df_2021[(df_2021["구분별(1)"] == "전체")].drop("구분별(1)", axis=1).reset_index(drop=True)
df_2021_all.rename(columns={"구분별(2)":"년도"}, inplace=True)
df_2021_all.iloc[0,0] = 2021
df_2021_all
```

```python
df_2019_all = df_2019[(df_2019["특성별(1)"] == "전체")].drop("특성별(1)", axis=1).reset_index(drop=True)
df_2019_all.rename(columns={"특성별(2)":"년도"}, inplace=True)
df_2019_all.iloc[0,0] = 2019
df_2019_all
```

```python
df_2017_all = df_2017[(df_2017["특성별(1)"] == "전체")].drop("특성별(1)", axis=1).reset_index(drop=True)
df_2017_all.rename(columns={"특성별(2)":"년도"}, inplace=True)
df_2017_all.iloc[0,0] = 2017
df_2017_all
```

```python
# 년도별 전체(소계) 데이터 행 단위 통합하기
df_2021_all.loc[len(df_2021_all)] = list(df_2019_all.iloc[0])
df_2021_all.loc[len(df_2021_all)] = list(df_2017_all.iloc[0])

# 년도를 int 타입으로 변환하기
df_2021_all["년도"] = df_2021_all["년도"].astype(int)
```

### 년도별 전체 전통시장 활성화 방안 : df_all

```python
import numpy as np
# 데이터
age_groups = list(df_2021_age["연령대"])
list1 = list(df_2021_age["주차장 건립 및 확장"])
list2 = list(df_2021_age["재래시장 시설 현대화"])
list3 = list(df_2021_age["거래형태 개선"])
list4 = list(df_2021_age["고객 편의시설 확충"])
list5 = list(df_2021_age["상인의 친절의식 혁신교육"])

# 선택할 가장 중요한 4가지
selected_indices = [0, 1, 2, 3, 4]  # 예시로 1, 2, 3, 4번째 항목을 선택하였습니다.

# 선택된 데이터 추출
selected_labels = ['주차장 건립 및 확장', '재래시장 시설 현대화', '거래형태 개선', '고객 편의시설 확충', '상인의 친절의식 혁신교육']
selected_data = [list1,list2,list3,list4,list5]

# 그래프 그리기
fig, ax = plt.subplots(figsize=(10,5.5))

bar_width = 0.15
opacity = 1
colors = ['#8cc7cf', '#61b299', '#cde59b', '#fdce7e' ,'#ffecb9']

for i in range(len(selected_indices)):
    x = np.arange(len(age_groups))
    x_shifted = x + (bar_width * i)
    ax.bar(x_shifted, selected_data[i], bar_width, alpha=opacity, color=colors[i], label=selected_labels[i],
           edgecolor='black', linewidth=1)
ax.set_title("2021년 연령별 전통시장 활성화 방안 설문조사 결과")
ax.set_xlabel('연령별')
ax.set_ylabel('비율(%)')
ax.set_xticks(x + (bar_width * (len(selected_indices) - 1) / 2))
ax.set_xticklabels(age_groups, rotation=45)
ax.legend()
wedgeprops={'width': 0.7, 'edgecolor': 'b', 'linewidth': 1}


plt.tight_layout()
plt.savefig('2021년 연령별 전통시장 활성화방안 설문조사 결과.png')
```
