```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

plt.rc("font", family = "Malgun Gothic")
```

## 데이터 불러오기

「부산광역시사회조사」 재래시장 활성화 방안 설문조사의 2021, 2019, 2017년 데이터를 활용하여 부산 시민들이 전통시장에서 어떤 점이 개선되었으면 하는지, 어떤 걸 선호하는지 등을 파악


```python
df_2017 = pd.read_csv("./data/재래시장_활성화_방안_2017.csv", encoding="euc-kr")
df_2019 = pd.read_csv("./data/재래시장_활성화_방안_2019.csv", encoding="euc-kr")
df_2021 = pd.read_csv("./data/재래시장_활성화_방안_2021.csv", encoding="euc-kr")
```

## 연령별 활성화 방안

### 데이터 전처리

```python
# 2021년 데이터에서 연령별 데이터만 추출
df_2021_age = df_2021[(df_2021["구분별(1)"] == "연령별")].drop("구분별(1)", axis=1).reset_index(drop=True)
df_2021_age.rename(columns={"구분별(2)":"연령대"}, inplace=True)
df_2021_age
```

```python
df_2019_age = df_2019[(df_2019["특성별(1)"] == "연령별")].drop("특성별(1)", axis=1).reset_index(drop=True)
df_2019_age.rename(columns={"특성별(2)":"연령대"}, inplace=True)
df_2019_age
```

```python
df_2017_age = df_2017[(df_2017["특성별(1)"] == "연령별")].drop("특성별(1)", axis=1).reset_index(drop=True)
df_2017_age.rename(columns={"특성별(2)":"연령대"}, inplace=True)
df_2017_age
```

```python
df2021 = df_2021_age.set_index("연령대")
df2021.T
```

### 데이터 시각화

예시로 2021년의 데이터만 시각화

```python
age_groups = ["15~19세", "20~29세", "30~39세", "40~49세", "50~59세", "60세 이상"]
plt.figure(figsize=(15, 7))

for age_group in age_groups:
    plt.plot(df2021.T[age_group], marker="o", linewidth=2, label=age_group)

plt.title("2021년 연령별 전통시장 활성화 방안", fontsize=15)
plt.xticks(rotation=15)
plt.grid()
plt.legend()

plt.tight_layout()
plt.savefig('./savefig/2021년_연령별_전통시장_활성화방안.png')
```

30세 이상부터는 '주차장 건립 및 확장'을 활성화 방안으로 많이 선택함<br>
반면, 어린 나이대에서는 '재래시장 시설 현대화'나 '거래형태 개선'을 선택함

흥미로운 점은 60세 이상 연령층에서 '재래시장 시설 현대화'가 활성화 방안으로 높은 선택을 받음

```python
# 데이터
age_groups = list(df_2021_age["연령대"])
selected_labels = ['주차장 건립 및 확장', '재래시장 시설 현대화', '거래형태 개선', '고객 편의시설 확충', '상인의 친절의식 혁신교육']
selected_data = [list(df_2021_age[label]) for label in selected_labels]

# 그래프 그리기
fig, ax = plt.subplots(figsize=(10, 5.5))

bar_width = 0.15
opacity = 1
colors = ['#8cc7cf', '#61b299', '#cde59b', '#fdce7e', '#ffecb9']

for i, (label, data) in enumerate(zip(selected_labels, selected_data)):
    x = np.arange(len(age_groups))
    x_shifted = x + (bar_width * i)
    ax.bar(x_shifted, data, bar_width, alpha=opacity, color=colors[i], label=label,
           edgecolor='black', linewidth=1)

ax.set_title("2021년 연령별 전통시장 활성화 방안")
ax.set_xlabel('연령별')
ax.set_ylabel('비율(%)')
ax.set_xticks(x + (bar_width * (len(selected_labels) - 1) / 2))
ax.set_xticklabels(age_groups, rotation=45)
ax.legend()

plt.tight_layout()
plt.savefig('./savefig/2021년_연령별_전통시장_활성화방안_막대.png')
```

막대 그래프를 통해 각 연령대가 어떤 활성화 방안을 선호하는지 명확하게 시각화함

## 년도별 활성화 방안

```python
# 2021년
df_2021_all = df_2021[(df_2021["구분별(1)"] == "전체")].drop("구분별(1)", axis=1).reset_index(drop=True)
df_2021_all.rename(columns={"구분별(2)":"년도"}, inplace=True)
df_2021_all.iloc[0,0] = 2021
df_2021_all
```

```python
# 2019년
df_2019_all = df_2019[(df_2019["특성별(1)"] == "전체")].drop("특성별(1)", axis=1).reset_index(drop=True)
df_2019_all.rename(columns={"특성별(2)":"년도"}, inplace=True)
df_2019_all.iloc[0,0] = 2019
df_2019_all
```

```python
# 2017년
df_2017_all = df_2017[(df_2017["특성별(1)"] == "전체")].drop("특성별(1)", axis=1).reset_index(drop=True)
df_2017_all.rename(columns={"특성별(2)":"년도"}, inplace=True)
df_2017_all.iloc[0,0] = 2017
df_2017_all
```

```python
# 년도별 전체(소계) 데이터 행 단위 통합하기
df_2021_all.loc[len(df_2021_all)] = list(df_2019_all.iloc[0])
df_2021_all.loc[len(df_2021_all)] = list(df_2017_all.iloc[0])

# 년도를 int 타입으로 변환하기
df_2021_all["년도"] = df_2021_all["년도"].astype(int)
```

```python
# 전체 년도 데이터 df_all
df_all = df_2021_all.copy()
df_all
```

```python
# 년도를 인덱스로
df_all.set_index("년도", inplace=True)
df_all
```

### 년도별 전통시장 활성화 방안

```python
df_all.T.plot(figsize=(10,7),linewidth=2,marker="o")
plt.xticks(rotation=90)
plt.ylim([0,50])
plt.title("년도별 전통시장 활성화 방안", fontsize=12)
plt.ylabel("응답률(단위:%)")

plt.tight_layout()
plt.savefig('./savefig/년도별_전통시장_활성화방안.png')
```

년도별로 확인해보아도 참여자들이 선호하거나 활성화하고자 하는 방안은 유사한 것으로 나타남

```python
def plot_survey_results(data, year):
    labels = list(data.columns)
    colors = ['#61b299', '#cde59b', '#fdce7e', '#8cc7cf', '#ffecb9', '#fd8c82', '#d395d0', '#588195', '#438820']
    wedgeprops = {'width': 0.7, 'edgecolor': 'w', 'linewidth': 4}

    plt.figure(figsize=(8, 8))
    plt.axis('equal')
    plt.title(f'{year}년 전통시장 활성화 방안 설문조사', size=15)

    for i in range(len(data)):
        ratio = list(data.iloc[i, :])
        plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, colors=colors, wedgeprops=wedgeprops)
        plt.savefig(f'./savefig/{year}년_전통시장_활성화방안_설문조사.png')
        plt.clf()

# 데이터 생성 및 플로팅
years = [2021, 2019, 2017]
for i in range(len(years)):
    plot_survey_results(df_all.iloc[[i]], years[i])
```

년도별 설문조사 결과를 파이 차트로 한눈에 보기 쉽게 시각화함
